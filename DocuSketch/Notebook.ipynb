### **Task 1. Imports and data loading, examine data, summary info, and descriptive stats**

```python
from plotter import Plotter
import requests
import pandas as pd

#save file
url = 'https://ai-process-sandy.s3.eu-west-1.amazonaws.com/purge/deviation.json'
response = requests.get(url)
with open('deviation.json', 'wb') as f:
    f.write(response.content)
data = pd.read_json('deviation.json')
```

### **Task 2. Main function** 

```python
#start main function
plot_drawer = Plotter(json_path='deviation.json')
plot_paths = plot_drawer.draw_plots()

# display the path to the saved charts
for path in plot_paths:
    print(f'Plot saved at: {path}')
```
